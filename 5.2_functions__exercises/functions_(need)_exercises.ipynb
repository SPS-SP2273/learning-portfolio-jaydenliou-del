## Exercise 1

```python
def greeting(name):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    return f'Hello {name}!'
```

```python
greeting('Batman')
```

```python
greeting('John')
```

If the argument is true (where name == Batman), the return statement 'Hello Batman! So, nice to meet you!' exits the function. If the argument is not true, the if block is not executed and is skipped, with the next line being the return f'Hello {name}!'.

## Exercise 2

```python
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    if bmi < 18.5:
        print(f"Your BMI is UNDERWEIGHT at {bmi:.3f}")
    if bmi >= 18.5 and bmi <= 24.9:
        print(f"Your BMI is NORMAL WEIGHT at {bmi:.3f}")
    if bmi >= 25 and bmi <= 29.9:
        print(f"Your BMI is OVERWEIGHT at {bmi:.3f}")
    if bmi >= 30:
        print(f"Your BMI is OBESE at {bmi:.3f}")
    
```

```python
calculate_bmi(700, 1.9)
```

## Exercise 3

```python
def calculate_factorials(n):
    start = 1
    for i in range(1, n + 1): # 'n + 1' allows inclusion of the value of 'n'
        start *= i # 1 * 2 * ... (n + 1)

    if n < 0:
        start -= 1
        print(f"The value input: {n} cannot be negative!")
    
    return start
        
        
```

```python
calculate_factorials(3)
```

```python
calculate_factorials(-2)
```
