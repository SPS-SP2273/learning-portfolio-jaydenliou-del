## Exercise 1 (Celsisus to Farenheit or Kelvin)

```python
# Celisus to Kelvin
# K = C + 273.15

# Celsius to Fahrenheit 
# F = ((9/5) * C) + 32
```

```python
def convert_celsius(temperature_celsius, target_scale = 'Fahrenheit'): # setting default value of target_scale = 'Fahrenheit'
    celsius = float(temperature_celsius)
    
    if target_scale == "Kelvin":
        return celsius + 273.15
    elif target_scale == "Fahrenheit": # using elif ensures the target_scale are mutually exclusive events
        return ((9/5) * celsius) + 32
    
    else:
        return "Please input 2nd argument as 'Kelvin' or 'Fahrenheit' only!"
```

```python
convert_celsius(50)
```

```python
convert_celsius(50, 'Fahrenheit')
```

```python
convert_celsius(50, 'Kelvin')
```

```python
convert_celsius(50, 'Celsius')
```

## Exercise 2 (Fahrenheit to Celsius or Kelvin)

```python
# Fahrenheit to Kelvin
# K = C + 273.15
# K = ((F - 32) * 5/9) + 273.15

# Celsius to Fahrenheit 
# F = ((9/5) * C) + 32
# C = (F - 32) * 5/9
```

```python
def convert_fahrenheit(temperature_fahrenheit, target_scale = 'Celsius'):
    fahren = float(temperature_fahrenheit)

    if target_scale == 'Kelvin':
        return ((fahren - 32) * 5/9) + 273.15
    
    elif target_scale == 'Celsius':
        return (fahren - 32) * 5/9
    
    else:
        return "Please input 2nd argument as 'Kelvin' or 'Celsius' only!"
```

```python
convert_fahrenheit(20)
```

```python
convert_fahrenheit(30, 'Kelvin')
```

## Exercise 3 (General Temperature Conversion)

```python
def convert_temperature(temperature, source_scale, target_scale):
    temp = float(temperature)
    if source_scale == 'Celsius' and target_scale == 'Kelvin':
        return temp + 273.15
    elif source_scale == 'Celsius' and target_scale == 'Fahrenheit':
        return ((9/5) * temp) + 32
    elif source_scale == 'Fahrenheit' and target_scale == 'Celsius':
        return (temp - 32) * 5/9
    elif source_scale == 'Fahrenheit' and target_scale == 'Kelvin':
        return ((temp - 32) * 5/9) + 273.15
    elif source_scale == 'Kelvin' and target_scale == 'Celsius':
        return temp - 273.15
    elif source_scale == 'Kelvin' and target_scale == 'Fahrenheit':
        return 9/5 * (temp - 273.15) + 32
    else:
        return "Please input only 'Celsius', 'Fahrenheit', or 'Kelvin' in the blanks (temperature value, ___, ___) only!"
```

```python
convert_temperature(20, 'Kelvin', 'Celsius')
```

```python
convert_temperature(20, 'Kelvin', 'Fahrenheit')
```

```python
convert_temperature(20, 'Kelvin', '')
```
