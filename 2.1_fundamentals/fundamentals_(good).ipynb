# 4 Never compare floats directly

## 4.1 The Problem

```python
a = 0.1
a3 = 0.3
a * 3 == a3
# Mathematically, the output should show True but the computer stores these small fractions in a way that it is slightly off by a tiny fraction
```

```python
f'{0.3:.17f}'
```

## 4.2 A solution

```python
eps = 1E-10
abs(a * 3 - a3) < eps
```

```python
import numpy as np
np.isclose(a * 3, a3)
```

## 6.1 Self-documenting f-strings

```python
x, y = 42, 24
print(f'{x=} and {y=}')
```

```python
x, y = 24/5, 24/5
print(f'{x=:.3f} and {y=:.6f}')
```

```python
print(f'{10 * 5 + 2=}') # for expressions & calculations
```

```python
print(f'{2**10=}') # for expressions & calculations
```

```python
name = 'Batman'
print(f'{name.title()}') # for string without = sign
```

```python
name = 'Batman'
print(f'{name.title()=}') # for string with = sign, shows the source code
```

```python
# Lists
items = ['apple', 'banana']
print(f'{items[0]=}') # items[0] accesses the first element in the list
```

```python
# Lists
items = ['apple', 'banana']
print(f'{items[1]=}')   # items[1] accesses the second element in the list
```

```python
# Dictionaries 
data = {'id': 101}
print(f'{data["id"]=}') # does not use a number (in list) but use a specific key
                        # data["id"] tells Python to find the value associated with the key "id"
```

# 7 Computers read = from Right to Left!

```python
x = 40  # the computer takes the value 40 and assigns it to the label x
y = x + 2 # the computer first calculates x + 2 (which is 42) and then assigns the final result to y
print(y)
```

```python
y = 40
y = y + 2   # In math this equation is impossible as y cannot be equal to itself plus 2
            # But the computer reads the current value of y (which is 40) and adds 2 to get 42 and assigns and updates the variable y
print(y)
```
