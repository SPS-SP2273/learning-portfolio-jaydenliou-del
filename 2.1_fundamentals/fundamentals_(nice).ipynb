# 1 If if is not enough

```python
# if - elif - else statements is cumbersome --> use a match - case statement 

name = 'Batman'

match name:
    case 'Batman':
        print('Hello Hero | Batman!')
    case 'Robin':
        print('Hello Sidekick | Robin!')
    case _:
        print('Hello World!')
```

```python
# if - elif - else statements is cumbersome --> use a match - case statement 

name = 'Jay'

match name:
    case 'Batman':
        print('Hello Hero | Batman!')
    case 'Robin':
        print('Hello Sidekick | Robin!')
    case _: # wildcard for "catch all"
        print('Hello World!')
```

```python
# Sample Use Case: Amino Acid Classifier

# use the pipe symbol "|" to group multiple patterns together

def classify_amino_acid(short_code):
    
    short_code = short_code.upper() # convert input to uppercase

    match short_code:
        case 'G' | 'A' | 'V':
            return "non-polar"
    match short_code: 
        case 'S' | 'T' | 'Y':
            return "polar (uncharged)"
```

```python
classify_amino_acid('s')
```

# 2 Ternary operators or Conditional Statements

```python
nationality = input("Please enter your nationality:")
if nationality == 'French':
    greeting = 'Bonjour!'
else:
    greeting = 'Hello!'

print(greeting)
```

```python
greeting = 'Bonjour!' if nationality == 'French' else 'Hello!'
print(greeting)
```

# 3 Swapping values

```python
# swapping values / tuple unpacking

a, b = 1, 2
a, b = b, a # python first looks at the right side then assigns based on the left side
print(a, b)
```

```python
# coordinate transformations

atom_pos = (12.5, -4.2, 30.1)

# tuple unpacking
x, y, z = atom_pos

# swapping
x, y = y, x

print(f"New coordinates: x = {x}, y = {y}, z = {z}")
```

```python
# Motorcycle Maintenance

last_service = 55000 
current_service = 50000

if last_service > current_service:
    last_service, current_service = current_service, last_service

print(f"Interval: {current_service - last_service}")
```
