## More Example with Math package

```python
#  degree and radian

import math

# convert degrees to radians 
angle_rad = math.radians(180)

# computing sine, cosine, and tangent 
val = math.sin(angle_rad)

print(f"sin(180 degree): {val}")
```

```python
# logarithms and exponents 

# pH calculation: pH = -log10[H+] / 1e-7 is a shorthand for 1.0 x 10^{-7}
h_ion_concentration = 1e-7
ph_level = -math.log10(h_ion_concentration)
print(ph_level)

# eulter's number (e) and powers
growth = math.exp(2) # same as e^2
print(growth)
```
