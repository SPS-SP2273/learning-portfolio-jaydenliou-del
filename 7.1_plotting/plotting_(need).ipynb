## 1.2 Adding another plot

```python
from matplotlib import pyplot as plt 
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label = 'Y values', c = 'red', ls = 'dashed', lw = 7)
plt.plot(x, y_2, label = 'Y2 values', c = 'blue', ls = 'solid', lw = 7)


plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title ('X vs Y')

plt.grid(alpha = .25) # alpha = .25 makes the grid lines semi-transparent
plt.legend(loc='upper left')
```

```python
# Simple Bar Chart

import matplotlib.pyplot as plt

categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
quantities = [10, 15, 7, 12]

plt.bar(categories, quantities, color = 'teal')

plt.xlabel('Fruit Type')
plt.ylabel('Quantity')
plt.title('Fruit Inventory')

# plt.savefig('simple_bar_chart.png') - this line saves into the folder this ipynb is located
```

```python
# Sales Growth Line Plot

import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
product_a = [10, 25, 18, 30, 42]
product_b = [15, 20, 35, 38, 32]

plt.figure(figsize=(10, 5))

# Plot Prodduct A: Solid blue line with circle markers
plt.plot(months, product_a, label = 'Tech Gadgets', c = 'royalblue', marker = 'o', lw = 2)

# Plot Product B: Dashed orange line with square markers
plt.plot(months, product_b, label = 'Home Decor', c = 'darkorange', marker = 's', ls = '--')

# Customizing the view
plt.title('Monthly Sales Performace', fontsize = 14)
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.ylim(0,50)  # Set the y-axis range to start at 0
plt.grid(axis = 'y', ls ='--', alpha = 0.7) # Grid lines for horizontal axis only / Remove axis = y completely to have all grids
plt.legend(loc = 'upper left')

plt.show()
```
