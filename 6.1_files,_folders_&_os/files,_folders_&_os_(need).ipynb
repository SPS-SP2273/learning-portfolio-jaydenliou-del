## 1.3 Path separator

Windows uses backslash "\\" in the path. Example: C:\Users\Jayden\Documents\data.csv

macOS / Linux uses forward slash "/". Example: /Users/Jayden/Documents/data.csv

The difference in OS environment will cause the code to break once the code is shared across environment. 

```python
# The solution: Using the os module, os.path.join()

# This might fail on Windows
path = "Documents/Github/project "

# Solution

import os

# Python will use / on Mac and \ if move to Windows
path = os.path.join("Documents", "Github", "project")
print(path)
```

## 1.4 Text files vs.Â Binary files

Text Files: plain text data that can be opened using textedit or notepad. Examples are ".txt", ".md", and ".csv". The downside is that they can become bulky as they aren't compressed.

Binary Files: complex files that require specific software to decode them. If opened using a text editor, you will see gibberish. Examples are ".png". They are OS specific - Excel.exe works on Windows while Excel.app works on Mac. Benefit is they are optimized for speed and take up less space than text files.

## 1.5 Extensions

Files are named with an extension at the end (name.extension). The extension tells the OS which app to use to read the file. For example, ".xlsx" extension tells the OS to use Excel.

## 2.2 Writing data

### Writing to a file in one go

```python
text = 'Far out in the uncharted backwaters of the unfashionable end of the western spiral arm of the Galaxy lies a small unregarded yellow sun.\nOrbiting this at a distance of roughly ninety-two million miles is an utterly insignificant little blue green planet whose ape-descended life forms are so amazingly primitive that they still think digital watches are a pretty neat idea.'
```

```python
with open('my-text-once.txt', 'w') as file:
    file.write(text)

# this creates a file 'my-text-once.txt' and 'w' indicates that I am opening the file for writing
# "with" statement is the context manager - ensure that the file is automatically closed as soon as the code inside the indented block finishes running
```

### Writing to a file, line by line

```python
with open('my_text_once.txt', 'a') as file:
    for line in text.splitlines():
        file.write(line)
```

# 5 Folders

## 5.1 Creating folders

```python
import os

os.mkdir('people')

for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    print(f'Creating {path}')
    os.mkdir(path)

# this creates a golder called 'people' in the current directory and subfolders 'John', 'Paul', 'Ringo'
```

## 5.2 Checking for existence

### Using try-except

```python
for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    try:
        os.mkdir(path)
        print(f'Creating {path}')
    except FileExistsError:
        print(f'{path} already exists; skipping creation.')
```

### Using os.path.exists()

```python
for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    if os.path.exists(path):
        print(f'{path} already exists; skipping creation.')
    else:
        os.mkdir(path)
        print(f'Creating {path}')
```

## 5.3 Copying files

```python
import os
import shutil

for person in ['John', 'Paul', 'Ringo']:
    path_to_destination = os.path.join('people', person)
    shutil.copy('sp2273_logo.png', path_to_destination) # this looks for the file name and creates a copy inside the specific folder
    print(f'Copied file to {path_to_destination}')
```

```python
for person in ['John', 'Paul', 'Ringo']:
    # Create folder 'imgs'
    path_to_imgs = os.path.join('people', person, 'imgs')
    if not os.path.exists(path_to_imgs): # this asks if the folder already exist
        os.mkdir(path_to_imgs)

    # Move logo file
    current_path_of_logo = os.path.join('people', person, 'sp2273_logo.png')
    new_path_of_logo = os.path.join('people', person, 'imgs', 'sp2273_logo.png')

    shutil.move(current_path_of_logo, new_path_of_logo)
    print(f'Moved logo to {new_path_of_logo}')
```

# 6 Listing and looking for files

```python
import glob

glob.glob('*')

# glob gives me anything in the folder, with '*' called a wildcard
```

```python
glob.glob('peo*')

# refine the search by matching the pattern 'peo' followed by 'anything'
```

```python
glob.glob('peo*/*')

# i want to know what is inside this folder that start with 'peo'
```

```python
glob.glob('people/**', recursive=True)

# 'people/**' - look inside the people folder and every single subfolder nested beneath it
# recursive = True is the on switch for the double asterisk.
# Without this, the '**' acts just like a single '*' and wouldn't go deep into the subdirectories
```

```python
glob.glob('people/**/*.png', recursive=True)

# '/*.png' is the filter saying "I only want files that end with the .png extension"
# recursive = True is mandatory if using **
```

```python
# ANIMATION CODE FROM AI 
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.animation as animation

# 1. Setup data
x = np.linspace(0, 2*np.pi, 100) # Creates a list of 100 numbers starting from 0 to 2pi for the x-axis
fig, ax = plt.subplots() # fig is the canvas and ax is the graph area

# Use a comma after 'line' to "unpack" the list returned by ax.plot
line, = ax.plot(x, np.sin(x), 'g')

# 2. Update function
def update(frame):
    # Using a smaller multiplier (0.1) so the wave moves smoothly
    y = np.sin(x - 0.1 * frame) # By subtracting a bit of the frame from x, we shift the wave to the right
    line.set_ydata(y)  # Use set_ydata for updating the vertical points
    return line,

# 3. Create the animation object
# We assign it to 'ani' so we can save it later
ani = animation.FuncAnimation(
    fig=fig, 
    func=update, 
    frames=100,      # Increased frames for a longer animation
    interval=20,     # Time between frames in milliseconds
    blit=True        # Helps performance
)

# 4. Save the animation
# Note: You may need 'pillow' installed (pip install pillow) to save as GIF
ani.save('my_animation.gif', writer='pillow')

plt.show()
```
