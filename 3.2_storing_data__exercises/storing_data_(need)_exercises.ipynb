## Exercise 5

```python
x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]
```

```python
# 1. Determine the means of x1 and x2
import numpy as np

np_x_1 = np.array(x_1)
np_x_2 = np.array(x_2)
print(np.mean(np_x_1))
print(np.mean(np_x_2))

```

```python
# Determine the standard deviations of x1 and x2

print(np.std(np_x_1))
print(np.std(np_x_2))
```

```python
# Determine the Pearson correlation coefficient r for the two datasets
x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]
np_x_1 = np.array(x_1)
np_x_2 = np.array(x_2)

# Calculate mean
mean_x_1 = np.mean(np_x_1)
mean_x_2 = np.mean(np_x_2)

# Covariance
cov = np.cov(np_x_1, np_x_2)
cov_2 = cov[0,1]
print(f"covariance = {cov_2:.4f}")

# By default np.std uses ddof=0, thus changing to ddof=1 is needed
std_x_1 = np.std(np_x_1, ddof=1)
std_x_2 = np.std(np_x_2, ddof=1)

r = cov_2 / (std_x_1 * std_x_2)
print(f"r = {r}")


```

```python
# Compare resuls to np.corrcoef()

r = np.corrcoef(np_x_1, np_x_2)
print(r)
```

```python
# Compare resuls to np.corrcoef()

matrix = np.corrcoef(np_x_1, np_x_2)
r = matrix[0, 1]

print(f"r = {r:.10f}")
```

```python
# Compare resuls to np.corrcoef()

matrix = np.corrcoef(np_x_1, np_x_2)
r = matrix[1, 0]

print(f"r = {r:.10f}")
```

When manually calculating the r value using covariance and standard deviation function, changing of the delta degree of freedom is needed in order for both function to have the same ddof value. Without changing the ddof value, the r value is > 1 which is invalid. When using np.corrcoef(), the output is a matrix which the r value is indexed at [1,0] and [0,1]. Similarly, when using np.cov(), the output is also in a matrix. Using np.corrcoef() is more efficient to calculate the r value.
