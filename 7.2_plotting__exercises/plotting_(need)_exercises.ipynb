## Exercise 1

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

import matplotlib.pyplot as plt 

plt.errorbar(x, y, 
             xerr= x_err, yerr= y_err,
             color= 'darkred', linestyle= '--',
             marker= 'o', markerfacecolor='yellow',
             ecolor= 'black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generated Data
x = np.linspace(0, 51, 100)
y = np.exp(-x/10) * np.sin(x)

# Plot
plt.plot(x, y, ls = '-', c = 'blue')
plt.title('$y = e^{-x/10} \sin(x)$', fontsize = 20)
plt.xlabel('x', c = 'darkorange', fontsize = 15)
plt.ylabel('y', c = 'darkorange', fontsize = 15, rotation = 0)
plt.grid(alpha=.25)
plt.ylim(-1, 1)
plt.show()
```
