```python
import glob
import shutil
import os

cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

folder_dir = './os-collaboration-exercise-data'

for city in cities:
    city_filter = f"*{city.lower()}.txt" # filters for city name in lowercase 
    files_move_list = glob.glob(os.path.join(folder_dir, city_filter))
    
    destination_folder = os.path.join('./Task 5/', city)

    for file_path in files_move_list:
        try:
            shutil.move(file_path, destination_folder)
            print(f"Moved {file_path} to {destination_folder}")
        except shutil.Error as error: # prevents the code from crashing and prints the error
            print(f"File already exits: {error}")
```
