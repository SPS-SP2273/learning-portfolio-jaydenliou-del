## 1.2 Anonymous functions

```python
my_short_function = lambda name: f"Hello {name}!"

my_short_function(name = 'Superman')
```

```python
numbers=[[9, 0, -10],
         [8, 1, -11],
         [7, 2, -12],
         [6, 3, -13],
         [5, 4, -14],
         [4, 5, -15],
         [3, 6, -16],
         [2, 7, -17],
         [1, 8, -18],
         [0, 9, -19]]

```

```python
sorted(numbers)

# The default sort: it looks at the first element of each sub-list
# logic: sort by x[0]
```

```python
sorted(numbers, key=lambda x: x[1])
# The column 2 sort
# the lambda function here tells Python: "ignore the first number, only look at the element at index 1"
# the result is the original list with 0 being the smallest in the 2nd column and 9 the largest in the 2nd column
```

```python
sorted(numbers, key=lambda x: sum(x))

# lambda calculates the total sum of each sub-list before deciding where to place it in the final list
# logic: sort by the result of value1 + value2 + value3
# first element:sum is -10
# last element: sum is -1
```

## 1.3 Optional arguments

```python
def greeting(name='no one'):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    else:
        return f'Hello {name}!'
```

```python
greeting()
```

```python
# Using default values
print('I', 'am', 'Batman!')
# Specifying an optional argument
print('I', 'am', 'Batman!', sep='---')  
```
