## 1.2 for with enumerate

```python
super_names = ["Black Widow", "Iron Man", "Doctor Strange"]
real_names = ["Natasha Romanoff", "Tony Stark", "Stephen Strange"]
```

```python
for count, name in enumerate(real_names):
    print(f"{count}: {name} is a Marvel superhero!")

# enumerate(real_names) returns pairs of (index, value) in this case (count, name) for each item in real_names
```

```python
for index, name in enumerate(real_names):
    superhero_name = super_names[index] # if we are at index 1(Tony Stark), it grabs index 1 from the superhero list (Iron Man)
    print(f"{name} is {superhero_name}!")
```

```python
# Application: Student Grades
students = ["Alice", "Bob", "Charlie"]
scores = [85, 58, 92]

for index, name in enumerate(students):
    # Use the index to grab the corresponding score
    score = scores[index]

    # Determine status
    status = "Passed" if score >= 60 else "Failed"

    print(f"Student: {name} | Score: {score} | Result: {status}")
```

```python
# The Structure
# The Loop: for index, item in enumerate(list_a):
#   index: the counter (starts at 0)
#   item: the actual data from list_a
# The Lookup: other_data = list_b[index]
#   This reaches into the second list at the exact same position the loop is currently at in the first list
```

```python
for count, name in enumerate(real_names, 100):
    print(f"{count}: {name} is a Marvel superhero!")
```

```python
# The Dictionary Approach
hero_registry = {
    "Natasha Romanoff": "Black Widow",
    "Tony Stark": "Iron Man",
    "Stephen Strange": "Doctor Strange"
}

# We iterate through the dictionary items
for real_name, super_name in hero_registry.items():
    print(f"{real_name} is {super_name}!")
```

## 1.3 for with range

```python
for i in range(5):
    print(i)
```

```python
for i in range(5,10): # 5 is the start value (inclusive) and 10 is the stop value (exclusive)
    print(i)
```

```python
for i in range(1, 10, 3): # 1 is the start value, while 10 is excluded but adds 3
    print(i)
```

```python
for i in range(len(real_names)):
    real_name = real_names[i]
    super_name = super_names[i]
    print(f"{real_name} is Marvel's {super_name}!")

# len(real_names): counts how many items are in the list
# range(3): creates a sequence of numbers - 0, 1, 2
```

# 2 while

```python
number = 0

while number < 5:
    print(number)
    number += 1

# number = 0 is the initialization
# while number < 5: is the condition
# number += 1 is the increment. Without this line, the number would stay 0 forever
```

```python
power_level = 10
goal = 50

while power_level < goal:
    power_level += 10
    if power_level == 30:
        break
    print(power_level)
```
