# 1 Interrupting the flow

```python
for power in range(5):
    number = 10**power
    print(power, number)
```

```python
for power in range(5):
    number = 10**power
    if number > 5000:
        break
    print(power, number)
```

```python
for power in range(5):
    if power == 3:  # where power is equal to 3, it tells Python: Stop what you are doing in this round and jump immediately back to the top for the next number
        continue    # any code written after the continue block is ignored for that specific round
    number = 10**power
    print(power, number)
```

```python
for number in range(10):
    if number % 2 == 0: # this checks if the number is EVEN
        continue    # stops the current iteration and jumps back to the top of the loop
    print(number)
```

```python
number = 0 # create a variable and assign it the value "0"

while True: # infinite loop: a "while loop" runs as long as its condition is "True"
    print(number)
    number += 1 # shorthand for number = number + 1
    if number > 4: break # the exit condition
```

```python
# more concise version of the "while loop"
number = 0

while number <= 4:
    print(number) # print first
    number += 1 # increment second
```

```python
# more concise version of the "while loop" (starting number is different as the print is below)
number = 0

while number <= 4:
    number += 1 # increment first
    print(number) # print second
```

```python
# User Input Validation (Common Use Case for while loop)
password = ""

while password != "secret123": # this means: keep running as long as the password is NOT equal to 'secret123'. If user types 'secret123', the condition becomes FALSE, and the loop exists
    password = input("Enter your password: ")
    if password != "secret123":
        print("Access Denied. Try again.")
        
print("Access Granted!")
```

# 2 List comprehension!

## 2.1 Basic syntax

```python
[number for number in range(5)]
# creating a list containing 0 to 4
# expression "number": is the output
# expression "for number"
```

```python
[number**2 for number in range(5)]
```

## 2.2 List comprehension with conditions

```python
# the LONG WAY
even_numbers = []
for number in range(10):
    if number % 2 == 0:
        even_numbers.append(number) # append() is to add a single item to the very end of an existing list
print(even_numbers)
```

```python
# the LIST COMPREHENSION WAY
[number for number in range(10) if number % 2 == 0]

# plain english: create a list of every number from 0 to 9, but only if that number is even
```

```python
[number*10 for number in range(10) if number % 2 == 0]
```

```python
# Data Transformation (Mapping)
celsius_temps = [0, 10, 20 , 30, 40]

# Convert C to F: (C * 9/5) + 32
farenheit_temps = [(temp * 9/5) + 32 for temp in celsius_temps] 

print(farenheit_temps)
```

```python
# String Cleaning & Formatting

raw_users = ["  alice ", "BOB", " charLIE  ", "dave"]

# Strip whitespace and capitalize the first letter

clean_users = [user.strip().capitalize() for user in raw_users]

print(clean_users)
```

# 3 Other useful stuff

## 3.1 for with unpacking

```python
x, y, z = [1, 2, 3]
print(f'x = {x}, y = {y}, z = {z}')
```

```python
# Use case: Swapping Variables
a = 10
b = 20
a, b = b, a # Now a is 20 and b is 10
print(a)
print(b)
```

```python
py_superhero_info = [['Natasha Romanoff', 'Black Widow'], ['Tony Stark', 'Iron Man'], ['Stephen Strange', 'Doctor Strange']]

for real_name, super_name in py_superhero_info:
    print(f"{real_name} is Marvel's {super_name}")

# py_superhero_info is a nested list, each inner list represents 2 elements - a Real Name and Superhero Name
# for loop: assigns Natasha Romanoff to real_name and Black Widow to super_name
```

```python
py_superhero_info = [['Natasha Romanoff', 'Black Widow'], ['Tony Stark', 'Iron Man'], ['Stephen Strange', 'Doctor Strange']]

for hero in py_superhero_info:
    print(f"{hero[0]} is Marvel's {hero[1]}")
```

## 3.2 for with zip()

```python
super_names = ["Black Widow", "Iron Man", "Doctor Strange"]
real_names = ["Natasha Romanoff", "Tony Stark", "Stephen Strange"]

for real_name, super_name in zip(real_names, super_names): # zip looks at index 0 of both lists and pairs them and so on.
    print(f"{real_name} is Marvel's {super_name}!")
```

## 3.3 for with dictionaries

```python
superhero_info = {"Natasha Romanoff": "Black Widow", "Tony Stark": "Iron Man", "Stephen Strange": "Doctor Strange"}

for key, value in superhero_info.items():
    print(f"{key} is Marvel's {value}!")
```

```python
for key in superhero_info.keys():
    value = superhero_info[key]
    print(f"{key} is Marvel's {value}!")
```
