## Exercise 1 

```python
temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50]

temperature_farenheit = [(c* 9/5) + 32 for c in temperatures_celsius]

for celsisus, farenheit in zip(temperatures_celsius, temperature_farenheit):
    print(f"{celsisus} degree C is {farenheit} degree F")
```

## Exercise 2

```python
for letter in ['A', 'B', 'C']:
    for number in [1, 2, 3]:
        print(f'{letter}{number}', end='\t')
    print('\n')
```

```python
for row in [1, 2, 3, 4, 5]:
    print(f"{row}:", end='\t')
    for column in [1, 2, 3, 4, 5]:
        product = row * column
        print(f"{product}", end = '\t')
    print('\n')

```
