## Exercise 1

```python
odd = []
for x in range(0,10):
    if x % 2 == 0:
        continue

    odd.append(x**2)
    print(odd)
```

```python
odd = []
for x in range(0,10):
    if x % 2 == 0:
        continue

    odd.append(x**2)
print(odd)
```

## Exercise 2

```python
odd = [ x**2 for x in range(0,10) if x % 2 == 1]
print(odd)
```

## Exercise 4

```python
units = 100
year = 0

while units > 50:
    units *= 0.95
    year += 1
    if units <50:
        break
    print(f"Year {year} is {units}")
```

```python
co2_data = [
    (2000, 369.55), (2001, 371.14), (2002, 373.28), 
    (2003, 375.80), (2004, 377.52), (2005, 379.80), 
    (2006, 381.90), (2007, 383.79), (2008, 385.60), 
    (2009, 387.43), (2010, 389.90), (2011, 391.65), 
    (2012, 393.85), (2013, 396.52), (2014, 398.65),
    (2015, 400.83), (2016, 404.24), (2017, 406.55), 
    (2018, 408.52), (2019, 411.44), (2020, 414.24)
]

prev_co2 = co2_data[0][1]
prev_year = co2_data[0][0]

for year, co2 in co2_data:
    diff_co2 = co2 - prev_co2
    diff_year = year - prev_year
    if diff_co2 >= 3 and diff_year == 1:
        print(f"From {prev_year} to {year}, CO2 increased by {diff_co2:.4f}.")
    
    prev_co2 = co2
    prev_year = year
```
